```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

H=np.arange( (6)*(12) ).reshape( (6 , 12  ) )
H=0*H
#print(H)
```

```python
#1367 emited from sun,  1050  direct,  1120 direct+indirect,   3.3% vari yearls
#   317 nekde mizi v atm.      70 indirect z atmosfery  247 premeny v atmosfere nebo ven 
#  pokud 70 ven,  177 zustava v atmosfere
#6,12  # INDICIES 0,0 DO NO WORK AT ALL
from math import sin,cos
for l in range(0,12):
    for w in range(1,6): # 0 doesnt work here
        #print(l,w)
        ang=abs(w-3)/2*3.1415/2
        H[w,l]=cos( ang )*(293-253) +253
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.imshow(H , cmap='spectral', interpolation='nearest', vmin=200, vmax=320)  #273
plt.axis('off')
extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
plt.savefig('map.png', facecolor='w', edgecolor='r', bbox_inches=extent, pad_inches=0)
```
