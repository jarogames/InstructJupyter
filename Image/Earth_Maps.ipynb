```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

H=np.arange(18/3*36/3).reshape( (18/3 , 36/3  ) )
H=0*H
#print(H)
fig = plt.figure()
ax = fig.add_subplot(111)
plt.imshow(H , cmap='spectral', interpolation='nearest', vmin=-35, vmax=600)
plt.axis('off')
extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
plt.savefig('map.png', facecolor='w', edgecolor='r', bbox_inches=extent, pad_inches=0)
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
map = Basemap(projection='ortho', lat_0 = 50, lon_0 = 0,
              resolution = 'l', area_thresh = 1000.)
map.warpimage( image='map.png' )
#map.drawrivers()
map.drawcoastlines()
#map.drawcountries()
#map.drawlsmask() # LAND
# draw lat/lon grid lines every 30 degrees.
map.drawmeridians(np.arange(0, 360, 20))
map.drawparallels(np.arange(-90, 90,20))
plt.show()
```
