```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
import numpy as np
left=2
top=150
top2=130
right=80
right2=120
bot=2
bot2=50
verts = [
    (left, bot), # left, bottom
    (left, top+50), # left, top
    (left+50, top+50), # left, top
    (left+50, top), # left, top
    (right, top ), # right, top
    (right, top2 ), # right, top
    (right2, top2), # right, bottom
    (right2, bot2), # right, bottom
    (right,  bot2), # right, bottom
    (right,  bot), # right, bottom
    (0., 0.), # ignored
    ]
codes = np.ones( len(verts), int) * Path.LINETO
codes[0]=Path.MOVETO
codes[-1]=Path.CLOSEPOLY
path = Path(verts, codes)
patch = patches.PathPatch(path, facecolor='none', lw=2)
fig, ax = plt.subplots( figsize=(10, 10), dpi=300 )
plt.gca().add_patch( patch )
mx,my=122,202
plt.axis( (0,mx, 0, my)  )
plt.axis('off')
extent = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
plt.savefig('pudorys.jpg', facecolor='w', edgecolor='r', bbox_inches=extent, pad_inches=0)
#plt.savefig('pudorys.jpg')
plt.show()
```
