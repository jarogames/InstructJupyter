```python
%matplotlib inline
def coord(p,plt):
    w=80
    h=40
    im1=plt.imread('/tmp/'+p+'.png')
    if p=='pi3B': 
        (x,y)=(100,380) 
    if p=='pibB':  
        (x,y)=(300,300) 
    if p=='pi4B':  
        (x,y)=(100,300) 
    if p=='piB':   
        (x,y)=(100,130) 
    if p=='pimB':  
        (x,y)=(400,10) 
    if p=='pix1B': 
        (x,y)=(500,10) 
    if p=='pix2B': 
        (x,y)=(600,10) 
    plt.imshow(im1,extent=(x,x+w,y,y+h),alpha=1.0)
    plt.text(x+w, y+h*1.5,p,style='italic', weight='bold', color='white',
    verticalalignment='bottom',horizontalalignment='right',
             bbox={'facecolor':'darkblue', 'alpha':1.0, 'pad':1})
    return

import matplotlib.pyplot as plt
img = plt.imread('pudorys.jpg')
mx,my=520,520
#plt.imshow(img ,  aspect='equal', alpha=0.9 ) # this is from top down:: plt.axis( (0,900, 210,0)  )
#plt.axis( (0,mx, 0, my)  )
#plt.axis( (0,600, 0,210) )  # flip image when without extent
#plt.axis('off')
im1=plt.imread('/tmp/pi3B.png')
plt.imshow(img ,  extent=(0,mx,0,my)  ,alpha=1.0 )
pis=[ 'pi3B','pi3B', 'pi4B', 'piB', 'pimB' ]
for i in pis:
    coord( i ,plt)
#plt.imsho
#w(im1  , extent=coord('pi3B')  ,alpha=1.0 )
plt.axis( (0,mx, 0, my)  )
plt.show()
```
